```python
# import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create

# Read in the data

df = pd.read_csv('movies.csv')
```

```python
# Let's look at the data

df.head()
```

```python
# Check for missing data

for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, percent_missing))
```

```python
# Data Types for Columns

df.dtypes
```

```python
# change data types for columns where appropriate

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
# create correct year column 

df['year_correct'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
# sort by gross revenue

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', 20)
```

```python
# Drop any duplicates

df.drop_duplicates()
```

```python
# Scatterplot with budget vs gross

plt.scatter(x=df['budget'],y=df['gross'])

plt.title('Budget vs. Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget',y='gross',data=df, scatter_kws={'color':'red'},line_kws={'color':'blue'})
```

```python
# Let's start looking at correlation

df.corr() 
```

```python
# Using heatmap to more clearly show correlation

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Create integer represenations of fields that contain non-integer data for deeper correlation analysis

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
# Use new integer representation fields to create new heatmap to check further correlations

correlation_matrix = df_numerized.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Use unstack() to clean up the new numerized data visually

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# Sort the correlations into pairs

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Sort the new correlation pairs by the highest correlation impact

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
